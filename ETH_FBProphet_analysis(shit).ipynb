```python
from fbprophet import Prophet
import datetime
import pandas_datareader.data as reader
import pandas as pd
```

```python
end = datetime.datetime.now()
start = datetime.date(end.year - 4, end.month, end.day)
```

```python
end = datetime.date(end.year - 1, end.month, end.day)
```

```python
start1 = datetime.date(end.year - 3, end.month, end.day)
```

```python
df = reader.get_data_yahoo('ETH-USD', start1, end)['Adj Close']
```

```python
df = pd.DataFrame(df)
df.reset_index(inplace=True)
```

```python
# We must name the column as "ds" for time and "y" for values
df_model = pd.DataFrame({'ds':df['Date'], 'y':df['Adj Close']})
```

```python
modelfb = Prophet()
```

```python
modelfb.fit(df_model)
```

```python
future = modelfb.make_future_dataframe(periods=365)
```

```python
forecast = modelfb.predict(future)
```

```python
modelfb.plot(forecast)
```

```python
future2 = modelfb.make_future_dataframe(periods=185)
forecast2 = modelfb.predict(future2)
modelfb.plot(forecast2)
```

```python
end = datetime.datetime.now()



start1 = datetime.date(end.year - 2, end.month, end.day)

df = reader.get_data_yahoo('ETH-USD', start1, end)['Adj Close']

df = pd.DataFrame(df)
df.reset_index(inplace=True)

# We must name the column as "ds" for time and "y" for values
df_model = pd.DataFrame({'ds':df['Date'], 'y':df['Adj Close']})

modelfb = Prophet()

modelfb.fit(df_model)

future = modelfb.make_future_dataframe(periods=365)

forecast = modelfb.predict(future)

modelfb.plot(forecast)
```
